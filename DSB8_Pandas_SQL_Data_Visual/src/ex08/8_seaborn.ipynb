## Dynamic of Commits per user in project1!

````python
import pandas as pd
import sqlite3
import seaborn as sbn
import matplotlib.pyplot as plt
````

````python
conn = sqlite3.connect('../data/checking-logs.sqlite')
````

````python
query="""
SELECT timestamp, uid, numTrials
FROM checker
WHERE status = 'ready' AND labname = 'project1' AND uid LIKE 'user%'; 
"""

checker = pd.read_sql_query(query, conn)
checker['timestamp']=checker['timestamp'].astype('datetime64[ns]')
checker
````

````python
plt.figure(figsize=(15, 10))

# Set background color to gray
plt.gca().set_facecolor('lightgray')

# Choose any palette you like
palette = sbn.color_palette("husl")  # you can change this

sbn.lineplot(
    data=checker,
    x='timestamp',
    y='numTrials',
    hue='uid',
    linewidth=3,
    palette=palette
)

plt.title("Dynamic of Commits per user in project1", fontsize=30)
plt.xlabel("timestamp", fontsize=15)
plt.ylabel("numTrials", fontsize=15)
plt.legend(bbox_to_anchor=(1.05, 0.5), loc='center left', borderaxespad=0.)
plt.tight_layout()
plt.show()
````

````python
conn.close()
````
