## 1. Preprocessing

1. Create the same dataframe as in the previous exercise.
2. Using `train_test_split` with parameters `test_size=0.2`, `random_state=21` get `X_train`, `y_train`, `X_test`, `y_test`. Use the additional parameter `stratify`.

````python
df=pd.read_csv('day-of-week-not-scaled.csv')
df.head()
````

````python
df2=pd.read_csv('dayofweek.csv')
df2.head()
````

````python
df['dayofweek']=df2['dayofweek']
df.head()
````

````python
y=df['dayofweek']<5
X=df.drop(columns=['dayofweek'])
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21, stratify=y)
````

## 2. SVM

1. Use the best parameters from the previous exercise and train the model of SVM.
2. You need to calculate `accuracy`, `precision`, `recall`, `ROC AUC`.

 - `precision` and `recall` should be calculated for each class (use `average='weighted'`)
 - `ROC AUC` should be calculated for each class against any other class (all possible pairwise combinations) and then weighted average should be applied for the final metric
 - the code in the cell should display the result as below:

```
accuracy is 0.88757
precision is 0.89267
recall is 0.88757
roc_auc is 0.97878
```

````python
def show_scores(model):
  acc_score=accuracy_score(y_test, model.predict(X_test))
  precision_val=precision_score(y_test, model.predict(X_test), average='weighted')
  recall_val=recall_score(y_test, model.predict(X_test), average='weighted')
  roc_auc=roc_auc_score(y_test, model.predict_proba(X_test)[:, 1], average='weighted')

  print("===\tSCORES\t===")
  print('accuracy is ', acc_score)
  print('precision is ', precision_val)
  print('recall is ', recall_val)
  print('roc_auc is ', roc_auc)
````

````python
svm=SVC(kernel='rbf', C=10, class_weight='balanced', gamma='auto', probability=True,
    random_state=21)
svm.fit(X_train, y_train)
````

````python
show_scores(svm)
````

## 3. Decision tree

1. The same task for decision tree

````python
decision_tree=DecisionTreeClassifier(criterion='gini', max_depth=22, random_state=21)
decision_tree.fit(X_train, y_train)
````

````python
show_scores(decision_tree)
````

## 4. Random forest

1. The same task for random forest.

````python
random_forest=RandomForestClassifier(class_weight='balanced', criterion='entropy',
                       max_depth=36, n_estimators=10)
random_forest.fit(X_train, y_train)
````

````python
show_scores(random_forest)
````

## 5. Predictions

1. Choose the best model.
2. Analyze: for which `weekday` your model makes the most errors (in % of the total number of samples of that class in your full dataset), for which `labname` and for which `users`.
3. Save the model.

````python
df_test = pd.concat([X_test, y_test], axis=1)

predictions = svm.predict(X_test)
df_test.rename(columns={'dayofweek': 'actual_is_weekday'}, inplace=True)

df_test['predicted_is_weekday'] = predictions


df_test['original_dayofweek'] = df.loc[X_test.index, 'dayofweek']
df_test['error'] = df_test['actual_is_weekday'] != df_test['predicted_is_weekday']

df_test.head()
````

````python
error_weekday=df_test.groupby('original_dayofweek')['error'].sum()/df_test.groupby('original_dayofweek')['error'].count()
error_weekday
````

````python
user_cols = [col for col in df_test.columns if col.startswith('uid_user_')]
user_error_rates = {}

for user_col in user_cols:
    user_data = df_test[df_test[user_col] == 1.0]
    if not user_data.empty:
        error_rate = user_data['error'].sum() / len(user_data)
        user_error_rates[user_col] = error_rate
    else:
        user_error_rates[user_col] = 0.0

users = pd.Series(user_error_rates).sort_values(ascending=False)
````

````python
lab_cols = [col for col in df_test.columns if col.startswith('labname_')]
lab_error_rates = {}

for lab_col in lab_cols:
    lab_data = df_test[df_test[lab_col] == 1.0]
    if not lab_data.empty:
        error_rate = lab_data['error'].sum() / len(lab_data)
        lab_error_rates[lab_col] = error_rate
    else:
        lab_error_rates[lab_col] = 0.0

labs = pd.Series(lab_error_rates).sort_values(ascending=False)
````

## 6. Function

1. Write a function that takes a list of different models and a corresponding list of parameters (dicts) and returns a dict that contains all the 4 metrics for each model.

````python
def model_evaluation(models, params):
  for model, param in zip(models, params):
    model.set_params(**param)
    model.fit(X_train, y_train)
    show_scores(model)
````

````python
models=[DecisionTreeClassifier(), RandomForestClassifier()]
params=[{'criterion':'gini', 'max_depth':22, 'random_state':21}, {'class_weight':'balanced', 'criterion':'entropy', 'max_depth':36, 'n_estimators':10}]
model_evaluation(models=models, params=params)
````
