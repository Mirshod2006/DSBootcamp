# MovieLens Data Analysis Report

This report analyzes the MovieLens dataset using custom classes from `movielens_analysis.py`. The analysis covers movies, ratings, tags, and links to IMDB data.

## Data Overview
- **Movies**: Information about films including titles, genres, and release years.
- **Ratings**: User ratings with timestamps.
- **Tags**: User-generated tags for movies.
- **Links**: Connections to IMDB and TMDB IDs.

## Data Loading

```python
from movielens_analysis import Ratings, Tags, Links, Movies

# Load data
ratings = Ratings("ratings.csv")
tags = Tags("tags.csv")
links = Links("links.csv")
movies = Movies("movies.csv")

print("Data loaded successfully.")
print(f"Movies: {len(movies.get_all())} entries")
print(f"Ratings: {len(ratings.get_all())} entries")
print(f"Tags: {len(tags.get_all())} entries")
print(f"Links: {len(links.get_all())} entries")
```

## Movie Analysis

Analyze the movies dataset for genre distributions, release years, and movies with the most genres.

```python
# Distribution by genres
genre_dist = movies.dist_by_genres()
print("Genre Distribution:")
for genre, count in list(genre_dist.items())[:10]:  # Top 10
    print(f"{genre}: {count}")

# Distribution by release year
release_dist = movies.dist_by_release()
print("\nRelease Year Distribution (Top 10):")
for year, count in list(release_dist.items())[:10]:
    print(f"{year}: {count}")

# Movies with most genres
most_genres = movies.most_genres(5)
print("\nMovies with Most Genres:")
for title, num in most_genres.items():
    print(f"{title}: {num} genres")
```

## Ratings Analysis

Explore ratings data: top films by average/median rating, controversial films, distribution by year, and films with most ratings.

```python
# Top films by average rating
top_avg = ratings.find_top_films(5, metric="average")
print("Top 5 Films by Average Rating:")
for title, rating in top_avg.items():
    print(f"{title}: {rating}")

# Top films by median rating
top_median = ratings.find_top_films(5, metric="median")
print("\nTop 5 Films by Median Rating:")
for title, rating in top_median.items():
    print(f"{title}: {rating}")

# Most controversial (highest variance)
controversial = ratings.top_controversial(5)
print("\nMost Controversial Films (by rating variance):")
for title, var in controversial.items():
    print(f"{title}: {var}")

# Distribution by year
year_dist = ratings.dist_by_year()
print("\nRating Distribution by Year:")
for year, count in sorted(year_dist.items()):
    print(f"{year}: {count}")

# Top by number of ratings
top_by_num = ratings.top_by_num_of_ratings(5)
print("\nFilms with Most Ratings:")
for title, num in top_by_num.items():
    print(f"{title}: {num} ratings")
```

## Tags Analysis

Analyze user tags: most popular, longest tags, and tags with most words.

```python
# Most popular tags
popular_tags = tags.most_popular(10)
print("Most Popular Tags:")
for tag, count in popular_tags.items():
    print(f"{tag}: {count}")

# Longest tags
longest_tags = tags.longest(5)
print("\nLongest Tags:")
for tag, length in longest_tags.items():
    print(f"{tag}: {length} characters")

# Tags with most words
most_words = tags.most_words(5)
print("\nTags with Most Words:")
for tag, words in most_words.items():
    print(f"{tag}: {words} words")
```

## IMDB Analysis

Fetch additional data from IMDB using links. Note: This requires internet access and may take time.

```python
# Example: Get IMDB info for a few movies
# Select a few movies from links
sample_movies = links.get_all()[:3]  # First 3
fields = ["Director", "Runtime", "IMDB Rating", "Budget", "Gross"]

imdb_data = links.get_imdb(sample_movies, fields)
print("IMDB Data for Sample Movies:")
for movie in imdb_data:
    print(movie)

# Top directors from the sample
if imdb_data:
    top_dirs = links.top_directors(imdb_data, 5)
    print("\nTop Directors:")
    for dir, count in top_dirs.items():
        print(f"{dir}: {count}")

# Most expensive from sample
if imdb_data:
    expensive = links.most_expensive(imdb_data, 5)
    print("\nMost Expensive Movies:")
    for title, budget in expensive.items():
        print(f"{title}: ${budget}")
```
