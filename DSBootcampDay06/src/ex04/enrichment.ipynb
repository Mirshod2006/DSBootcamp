## 2

```python
samples=auto_df.sample(n=200, random_state=21, replace=True)
samples
```

```python
samples['Fines']=auto_df['Fines'].sample(n=200, random_state=21, replace=True).values
samples['Refund']=auto_df['Refund'].sample(n=200, random_state=21, replace=True).values

print('Original shape: ', auto_df.shape)
auto_df= pd.concat([auto_df, samples], ignore_index=True)
print('Enriched shape : ', auto_df.shape)
```

## 4

```python
surnames=pd.read_json('../../datasets/surname.json', orient='records')

surnames.columns = surnames.iloc[0]
surnames=surnames[1:].reset_index(drop=True)
surnames
```

```python
np.random.seed(21)

# This approach is pretty complicted
surnames_series=pd.Series(surnames['NAME'].iloc[np.random.randint(0, surnames.shape[0], size=size)])

# previous but good approach
# surname_series = surname['NAME'].sample(n=size, replace=True, random_state=21).reset_index(drop=True)
surnames_series
```

```python
owners=pd.DataFrame()
surnames_series=surnames_series.reset_index(drop=True)
owners['CarNumber']=auto_df['CarNumber']
owners['SURNAME']=surnames_series
owners
```

```python
fines
```

```python
new_columns=pd.DataFrame({
    'CarNumber': ['A123BC161RUS', 'X777XX197RUS', 'M555MM163RUS', 'K900KK199RUS', 'H432HH154RUS'],
    'Make': ['Ford', 'Skoda', 'Toyota', 'Volkswagen', 'Ford'],
    'Model': ['Focus', 'Octavia', 'Camry', 'Passat', 'Fiesta'],
    'Fines': [3200, 15000, 6000, 8500, 2000],
    'Refund': [0, 1, 2, 0, 1],
    'Year': [2012, 2018, 2016, 2014, 2010]
})

print('Original shape: ', fines.shape)
fines= pd.concat([fines, new_columns], ignore_index=True)
print('Shape with new features : ', fines.shape)
```

```python
owners.drop_duplicates(subset='SURNAME', inplace=True)
owners
```

```python
owners=owners.iloc[:-20].copy()
owners
```

```python
three_new_cols = pd.DataFrame({
    'CarNumber':['Y163O8161RUS', 'E432XX77RUS', '92918M178RUS'],
    'SURNAME':['RICHARDSON', 'ROSS', 'LOPEZ']
})

print('Original shape: ', owners.shape)
owners=pd.concat([owners, three_new_cols],ignore_index=True)
print('Changed shape: ', owners.shape)
```

```python
df_inner = fines.merge(owners, on='CarNumber', how='inner')

df_outer=fines.merge(owners, on='CarNumber', how='outer')

df_left=fines.merge(owners, on='CarNumber', how='left')

df_right=fines.merge(owners, on='CarNumber', how='right')
```
