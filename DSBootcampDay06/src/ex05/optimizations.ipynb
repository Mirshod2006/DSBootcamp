```python
import pandas as pd
import numpy as np
import gc
import warnings
warnings.filterwarnings('ignore')
```

```python
fines_df=pd.read_csv('../ex04/fines.csv', sep=';')
fines_df
```

# 2

```python
%%timeit

new_fines_df_loop=fines_df.copy()

result=[]
for i in range(0, len(fines_df)):
    fines=new_fines_df_loop.iloc[i]['Fines']
    refund=new_fines_df_loop.iloc[i]['Refund']
    year=new_fines_df_loop.iloc[i]['Year']
    portion = (fines / refund) * year if refund != 0 else np.nan
    result.append(portion)

new_fines_df_loop['Fine_Portion']=result
```

```python
%%timeit

new_fines_df_iter=fines_df.copy()

values=[]

for _, row in fines_df.iterrows():
    portion= (row['Fines'] / row['Refund']) * row['Year'] if row['Refund'] != 0 else np.nan
    values.append(portion)

new_fines_df_iter['Fine_Portion']=values
```

```python
%%timeit

new_fines_df_apply = fines_df.copy()

new_fines_df_apply['Fine_Portion'] = new_fines_df_apply.apply(lambda row: 
                                (row['Fines'] / row['Refund']) * row['Year'] if row['Refund'] != 0 else np.nan,
                                axis=1)
```

```python
%%timeit

new_fines_df_series=fines_df.copy()

new_fines_df_series['Fine_Portion'] = (new_fines_df_series['Fines'] / new_fines_df_series['Refund']) * new_fines_df_series['Year']
```

```python
%%timeit

new_fines_df_values=fines_df.copy()

f=new_fines_df_values['Fines'].values
r=new_fines_df_values['Refund'].values
y=new_fines_df_values['Year'].values

new_fines_df_values['Fine_Portion'] = (f/r)*y
```
