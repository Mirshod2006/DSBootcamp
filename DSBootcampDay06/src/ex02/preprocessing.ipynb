```python
import pandas as pd
```

## 1

```python
auto_df=pd.read_csv("../data/auto.csv", index_col='ID')
auto_df
```

## 2

```python
auto_df.count()
```

## 3

```python
auto_df.drop_duplicates(subset=['CarNumber', 'Make_n_model', 'Fines'], keep='last', inplace=True)
auto_df.count()
```

## 4

```python
auto_df.isna().sum()
```

### History column is more than 500 missing values, so I must to drop duplicates!

```python
auto_df.drop(columns='History', inplace=True)
auto_df
```

```python
auto_df['Refund'] = auto_df['Refund'].ffill()
auto_df.isna().sum()
```

```python
auto_df['Fines'].fillna(auto_df['Fines'].mean(),inplace=True)
auto_df.isna().sum()
```

## 5

```python
auto_df[['Make', 'Model']] = auto_df['Make_n_model'].str.split(' ', expand=True)
auto_df.drop(columns='Make_n_model', inplace=True)
auto_df
```

```python
auto_df.to_json('auto.json',orient='records', indent=4)
auto_df.count()
```
