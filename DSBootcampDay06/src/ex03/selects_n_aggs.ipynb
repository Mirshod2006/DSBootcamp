```python
import pandas as pd
```

## 1

```python
auto_df = pd.read_json('../ex02/auto.json')
auto_df.set_index('CarNumber', inplace=True)
auto_df
```

## 2

```python
auto_df.loc[auto_df['Fines'] >= 2100]
```

```python
auto_df.loc[(auto_df['Fines'] >= 2100) & (auto_df['Refund'] == 2)]
```

```python
auto_df.loc[(auto_df['Model'] == 'Focus') | (auto_df['Model'] == 'Corolla')]
```

```python
car_numbers = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
auto_df.loc[auto_df.index.isin(car_numbers)]

# OR I CAN USE .apply() FUNCTION TO CHECK THIS CASE!
# auto_df.loc[auto_df['CarNumber']].apply(lambda x: x in car_numbers)
```

## 3

```python
auto_df.groupby('Make')['Fines'].median()
```

```python
auto_df.groupby(['Make', 'Model'])['Fines'].median()
```

```python
auto_df.groupby(['Make', 'Model'])['Fines'].value_counts()
```

```python
print('Minimum Values :')
print(auto_df.groupby(['Make', 'Model'])['Fines'].min())
print('Maximum Values :')
print(auto_df.groupby(['Make', 'Model'])['Fines'].max())
```

```python
print('Standard Deviation :')
print(auto_df.groupby(['Make', 'Model'])['Fines'].std())
```

## 4

```python
auto_df.groupby(auto_df.index)['Fines'].count().sort_values(ascending=False)
```

```python
top_car = auto_df.index.value_counts().index[0]

top_car_df = auto_df[auto_df.index == top_car]
top_car_df
```

```python
auto_df.groupby(auto_df.index)['Fines'].sum().sort_values(ascending=False)
```

```python
top_payer=auto_df.groupby(auto_df.index)['Fines'].sum().idxmax()
auto_df.loc[auto_df.index == top_payer]
```

```python
check_number=auto_df.groupby(auto_df.index)['Model'].nunique()

if len(check_number[check_number > 1 ]) > 0:
    print("Yes, there are car numbers belongs to different models:")
    print(check_number[check_number > 1 ])
else:
    print('No, there is no rule-breakers!')
```
